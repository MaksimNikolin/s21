```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect("../data/checking-logs.sqlite")
schema = pd.io.sql.read_sql("PRAGMA table_info(pageviews);", conn)
print(schema)
```

```python
pageviews_10 = pd.read_sql("SELECT * FROM pageviews LIMIT 10;", conn)
print(pageviews_10)
```

```python
pageviews = pd.read_sql("""
                SELECT uid, datetime
                FROM pageviews 
                WHERE uid LIKE 'user_%' 
                ORDER BY uid;
                """, conn, index_col='datetime')

pageviews.index = pd.to_datetime(pageviews.index)
```
