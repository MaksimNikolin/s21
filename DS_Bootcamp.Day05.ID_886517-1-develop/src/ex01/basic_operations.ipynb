```python
import pandas as pd

views = pd.read_csv('../feed-views.log',
                    delimiter='\t',
                    names=['datetime', 'user'], header=None,
                    )

views.shape
```

```python
views.dtypes
```

```python
views['datetime'] = views['datetime'].astype('datetime64[ns]')
```

```python
views.dtypes
```

```python
views['year']   = views['datetime'].dt.year
views['month']  = views['datetime'].dt.month
views['day']    = views['datetime'].dt.day
views['hour']   = views['datetime'].dt.hour
views['minute'] = views['datetime'].dt.minute
views['second'] = views['datetime'].dt.second

views.head(2)
```

```python
views = views.set_index('user')

bins = [0, 4, 7, 11, 17, 20, 24]
labels = ['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening']
views['daytime'] = pd.cut(views['hour'], bins=bins, labels=labels, include_lowest=True)

views.sample(2)
```

```python
views.count()
```

```python
views['daytime'].value_counts()
```

```python
views.sort_values(['hour', 'minute', 'second'], ascending=True, inplace=True)
```

```python
views.head()
```

```python
print(f"Min hour:", views['hour'].min())
print(f"Max hour:", views['hour'].max())
```

```python
print(f"Mode of daytime:", views['daytime'].mode()[0])
```

```python
print(f"Max hour where daytime is night:", views[views['daytime']=='night']['hour'].max())
print(f"Min hour where daytime is morning:", views[views['daytime']=='morning']['hour'].min())
```

```python
print(f"User where night is max:", views[(views['daytime'] == 'night') & (views['hour'] == views[views['daytime']=='night']['hour'].max())].head(1).index[0])
print(f"User where morning is min:", views[(views['daytime'] == 'morning') & (views['hour'] == views[views['daytime']=='morning']['hour'].min())].head(1).index[0])
```

```python
print(f"Mode of hour:", views['hour'].mode()[0])
print(f"Mode of daytime:", views['daytime'].mode()[0])
```

```python
print(f"3 users with smallest time in the morning:", views[views['daytime']=='morning']['hour'].nsmallest(3))
print(f"3 users with largest time in the morning:", views[views['daytime']=='morning']['hour'].nlargest(3))
```
