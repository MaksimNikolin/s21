```python
import pandas as pd

df = pd.read_json('../ex02/auto.json')
df.set_index('CarNumber', inplace=True)
df.head(2)
```

```python
print(df[df['Fines'] > 2100])
```

```python
print(df[(df['Fines'] > 2100) & (df['Refund'] == 2)])
```

```python
print(df[(df['Model'] == 'Focus') | (df['Model'] == 'Corolla')])
```

```python
number_list=['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
print(df[df.index.isin(number_list)])
```

```python
print(df.groupby('Make')[['Fines']].median())
```

```python
print(df.groupby(['Make', 'Model'])[['Fines']].median())
```

```python
print(df.groupby(['Make', 'Model'])[['Fines']].count())
```

```python
print(df.groupby(['Make', 'Model'])[['Fines']].min().astype(int))
```

```python
print(df.groupby(['Make', 'Model'])[['Fines']].max().astype(int))
```

```python
print(df.groupby(['Make', 'Model'])[['Fines']].std())
```

```python
print(df.groupby(df.index).size().sort_values(ascending=False))
```

```python
top1_car = df.groupby(df.index).size().sort_values(ascending=False).head(1).index.values[0]
print(df[df.index == top1_car]) 
```

```python
print(df[['Fines']].groupby(df.index).sum().astype(int).sort_values(['Fines'], ascending=False))
```

```python
top1_car = df[['Fines']].groupby(df.index).sum().astype(int).sort_values(['Fines'], ascending=False).head(1).index.values[0]
print(df[df.index == top1_car])
```

```python
df = df.reset_index()
print(df[df['CarNumber'].isin(df.groupby('CarNumber')['Model'].nunique()[lambda x: x > 1].index)])
```
