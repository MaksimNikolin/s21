```python
import pandas as pd

fines = pd.read_csv('../ex04/fines.csv')
```

```python
fines.head(2)
```

```python
def calc_loop(df):
    res = []
    for i in range(len(df)):
        value = df.iloc[i]['Fines'] / (df.iloc[i]['Refund'] * df.iloc[i]['Year'])
        res.append(value)
    return res

def calc_iterrows(df):
    res = []
    for _, row in df.iterrows():
        value = row['Fines'] / (row['Refund'] * row['Year'])
        res.append(value)
    return res
```

```python
%timeit fines['Calc_loop'] = calc_loop(fines)
```

```python
%timeit fines['Calc_iterrows'] = calc_iterrows(fines)
```

```python
%timeit fines['Calc_apply'] = fines.apply(lambda row: row['Fines'] / (row['Refund'] * row['Year']), axis=1)
```

```python
%timeit fines['Calc_series'] = fines['Fines'] / (fines['Refund'] * fines['Year'])
```

```python
%timeit fines['Calc_numpy'] = fines['Fines'].values / (fines['Refund'].values * fines['Year'].values)
```

```python
fines_optimized = fines.copy()

float_cols = fines_optimized.select_dtypes(include=['float64']).columns
fines_optimized[float_cols] = fines_optimized[float_cols].apply(pd.to_numeric, downcast='float')

int_cols = fines_optimized.select_dtypes(include=['int64']).columns
fines_optimized[int_cols] = fines_optimized[int_cols].apply(pd.to_numeric, downcast='integer')

fines_optimized.info(memory_usage='deep')
```

```python
object_cols = fines.select_dtypes(include='object').columns

for col in object_cols:
    fines[col] = fines[col].astype('category')

fines.info(memory_usage='deep')
```
