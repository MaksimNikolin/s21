```python
import pandas as pd

df = pd.read_csv('../auto.csv', index_col='ID')
df.head(2)
```

```python
df.count()
```

```python
df = df.drop_duplicates(['CarNumber', 'Make_n_model', 'Fines'], keep='last')
```

```python
df.count()
```

```python
df.isna().sum()
```

```python
df = df.dropna(axis=1, thresh=500)
df.isna().sum()
```

```python
df['Refund'] = df['Refund'].ffill()
df.isna().sum()
```

```python
df['Fines'] = df['Fines'].fillna(df['Fines'].mean())
df.isna().sum()
```

```python
df[['Make', 'Model']] = df['Make_n_model'].apply(lambda x: pd.Series(x.split(' ')))
df.drop(axis=1, columns=['Make_n_model'], inplace=True)
```

```python
df.to_json('auto.json')
```
